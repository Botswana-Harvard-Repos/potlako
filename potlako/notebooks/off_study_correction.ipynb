```python
import os
import sys

import django

sys.path.append('../..')  # add path to project root dir
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'potlako.settings')
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

# for more sophisticated setups, if you need to change connection settings (e.g. when using django-environ):
#os.environ["DATABASE_URL"] = "postgres://myuser:mypassword@localhost:54324/mydb"

# Connect to Django ORM
django.setup()
```

```python
from edc_visit_schedule.constants import ON_SCHEDULE
from edc_visit_schedule.models import SubjectScheduleHistory
from potlako_prn.models import *

exited_prts = CoordinatorExit.objects.filter(
    patient_disposition='exit')

for prt in exited_prts:
    try:
        schedule_history_obj = SubjectScheduleHistory.objects.get(
            subject_identifier=prt.subject_identifier,
        )
    except SubjectScheduleHistory.DoesNotExist:
        print(f'{prt.subject_identifier} does not exist in schedule history')
        pass
    else:
        schedule_history_obj.offschedule_datetime = None
        schedule_history_obj.schedule_status=ON_SCHEDULE
        schedule_history_obj.save()
    prt.patient_disposition = 'tag_for_endpoint_recording'
    prt.save()
    print(f'{prt.subject_identifier} updated')
```
